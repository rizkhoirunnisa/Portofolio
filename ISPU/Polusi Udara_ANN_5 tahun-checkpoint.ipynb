## BUILDING THE ARTIFICIAL NEURAL NETWORK

### IMPORTING Sequential DARI MODULE tensorflow

```python
from tensorflow.keras.models import Sequential
```

### IMPORTING Dense DARI MODULE tensorflow

```python
from tensorflow.keras.layers import Dense
```

### BUILD A NEURAL NETWORK

```python
np.random.seed(123)

model1 = Sequential()

#First Hidden Layer
model1.add(Dense(7, activation= 'elu', kernel_initializer = 'random_normal', input_dim = 6))

#Second  Hidden Layer
model1.add(Dense(7, activation = 'elu', kernel_initializer = 'random_normal'))

#Output Layer
model1.add(Dense(4, activation = 'softmax', kernel_initializer = 'random_normal'))
```

### COMPILING THE MODEL

```python
np.random.seed(123)

model1.compile(
    optimizer = 'adam', 
    loss = 'categorical_crossentropy', 
    metrics = ['accuracy']
)
```

### IMPORTING EarlyStopping DARI MODULE tensorflow

```python
from tensorflow.keras.callbacks import EarlyStopping
```

### Customizing Early Stopping

```python
np.random.seed(123)

custom_early_stopping = EarlyStopping(monitor = 'val_accuracy', patience = 10, min_delta = 0.001, mode = 'max')
```

### FITTING THE MODEL TO THE TRAINING DATA

```python
np.random.seed(123)

network = model1.fit(X_train, Y_train, epochs = 100, validation_split = 0.2, batch_size = 16, verbose = 2, 
                    callbacks = [custom_early_stopping])
```

```python
%config InlineBackend.figure_format = 'svg'

def plot_metric(network, metric):
    train_metrics = network.history[metric]
    val_metrics = network.history['val_' + metric]
    epochs = range(1, len(train_metrics) + 1)
    plt.plot(epochs, train_metrics)
    plt.plot(epochs, val_metrics)
    plt.title('Training and validation '+ metric)
    plt.xlabel('Epochs')
    plt.ylabel(metric)
    plt.legend(['train_' + metric, 'val_'+ metric])
    plt.show()
```

### PARAMETER TUNING USING KERAS TUNER

```python
np.random.seed(123)

def model_builder(hp):
    model = keras.Sequential()
    model.add(keras.layers.Flatten(input_shape=(6,)))

    # Tune the number of units in the first Dense layer
    hp_units = hp.Int('units', min_value=8, max_value=128, step=8)
    model.add(keras.layers.Dense(units=hp_units, activation='relu'))
    model.add(keras.layers.Dense(4,activation = 'softmax'))

    # Tune the learning rate for the optimizer
    # Choose an optimal value from 0.01, 0.001, or 0.0001
    hp_learning_rate = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])
    
    model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=hp_learning_rate), 
    loss = 'categorical_crossentropy', 
    metrics = ['accuracy'])
    
    return model
```

```python
np.random.seed(123)

tuner = kt.Hyperband(model_builder,
                     objective='val_accuracy',
                     max_epochs=100,
                     factor = 3,
                     directory='my_direct',
                     project_name='project')
```

```python
np.random.seed(123)

tuner.search_space_summary()
```

```python
np.random.seed(123)

stop_early = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=10)
```

```python
np.random.seed(123)

tuner.search(X_train, Y_train, epochs=50, validation_split=0.2,callbacks=[stop_early])

# Get the optimal hyperparameters
np.random.seed(123)

best_hps=tuner.get_best_hyperparameters(num_trials=1)[0]

print(f"""
The hyperparameter search is complete. The optimal number of units in the first densely-connected
layer is {best_hps.get('units')} and the optimal learning rate for the optimizer
is {best_hps.get('learning_rate')}.
""")
```

```python
# Build the model with the optimal hyperparameters and train it on the data for 50 epochs
np.random.seed(123)

model2 = tuner.hypermodel.build(best_hps)
history = model2.fit(X_train, Y_train, epochs=50, validation_split=0.2)

val_acc_per_epoch = history.history['val_accuracy']
best_epoch = val_acc_per_epoch.index(max(val_acc_per_epoch)) + 1
print('Best epoch: %d' % (best_epoch,))
```
