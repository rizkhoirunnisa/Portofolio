# 2.  Permen

## Pie Chart

```python
candy_names = ['Kit Kat', 'Snickers', 'Milky Way', 'Toblerone', 'Twix']
candy_counts = [52, 39, 78, 13, 78]
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D', '#D53032')
explode = (0,0.15,0,0,0)
```

```python
fig = plt.figure(figsize = (12, 8))


plt.pie(
 candy_counts,
 labels=candy_names,
 autopct='%1.1f%%',
 colors=colors,
 explode = explode,
 shadow=True)
plt.title('Probability of Candy', fontsize = '20', loc = 'right') 
plt.show()
```

1. Chart yang dipilih sesuai dengan problem 2 adalah Pie Chart, sebab dengan menggunakan Pie Chart kita dapat menunjukkan presentase setiap jenis permen berbanding keseluruhan permen, serta besaran peluangnya.

# 3. Makanan

```python
dessert_sales = {
 'Lava Cake': 14,
 'Mousse': 5,
 'Chocolate Cake': 12,
 'Ice Cream': 19,
 'Truffles': 6,
 'Brownie': 8,
 'Chocolate Chip Cookie': 12,
 'Chocolate Pudding': 9,
 'Souffle': 10,
 'Chocolate Cheesecake': 17,
 'Chocolate Chips': 2,
 'Fudge': 9,
 'Mochi': 13,
}
```

```python
desserts = list(dessert_sales.keys())
sales = list(dessert_sales.values())
df = pd.DataFrame({'Makanan': desserts, 'Jumlah': sales})
df.sort_values(by='Jumlah', inplace=True, ascending=False)
```

```python
fig = plt.figure(figsize = (10, 8))
x_coords = np.arange(len(df))
plt.bar(x_coords, df['Jumlah'], tick_label=df['Makanan'])
plt.xticks(rotation=90)
plt.ylabel('Popularity')
plt.title('Dessert Popularity',fontsize = '20', loc = 'center')
plt.show()
```

1. Chart yang dipilih sesuai dengan problem 3 adalah Bar Chart, sebab dari problem tersebut perlu diketahui 3 makanan penutup yang paling tidak populer. Itu artinya kita perlu mengetahui tingkat popularitas dari masing-masing makanan penutup yang ada di restoran tersebut berdasarkan tingkat penjualan dalam seminggu terakhir.
2. 3 Makanan penutup yang harus dikeluarkan dari menu adalah Tuffles, Mousse, dan Chocolate Chips
